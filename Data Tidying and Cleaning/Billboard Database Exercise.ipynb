### *Splitting the dataframe and indexing


```python
billboard_data = billboard_data.reset_index()
```

```python
song_data_columns = ["index", "year", "artist", "track", "time"]
song_data = billboard_data[song_data_columns]
```

```python
song_data
```

```python
history_columns = billboard_data.columns[~billboard_data.columns.isin(song_data_columns)]
```

```python
history = billboard_data[history_columns].reset_index()
```

```python
history.melt(id_vars = ["index", "date.entered"], var_name = "week", value_name = "position")
```
